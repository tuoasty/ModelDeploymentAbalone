```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
df.dropna()
```

```python
df["Sex"].unique()
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
df["Sex"] = encoder.fit_transform(df["Sex"])
```

```python
df.dtypes
```

```python
df["Sex"].unique()
```

```python
x = df[["Sex", "Length", "Diameter", "Height", "Whole weight", "Whole weight.1", "Whole weight.2", "Shell weight"]]
y = df["Rings"]
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
print("R^2 Score:", r2_score(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
```

```python
from sklearn.ensemble import GradientBoostingRegressor

gbr = GradientBoostingRegressor(n_estimators=200, learning_rate=0.1, max_depth=5, random_state=42)
gbr.fit(x_train, y_train)

y_pred = gbr.predict(x_test)

print("R^2 Score:", r2_score(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
```

```python
import joblib

joblib.dump(gbr, "model.pkl")
joblib.dump(encoder, "encoder.pkl")
```
